```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
type(tips)
```

```python
sns.distplot(tips['total_bill'], kde=False, bins=40)
```

```python
sns.pairplot(tips, hue='sex', palette = 'coolwarm')
```
